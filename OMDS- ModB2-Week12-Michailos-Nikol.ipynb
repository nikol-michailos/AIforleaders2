# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```


### 📊 Weekly Graph Question – Advantages & Disadvantages

**Line chart (left panel):**
- **Advantages:** Shows overall trend clearly, connects data points across the week, easy to see rises and dips.
- **Disadvantages:** Overlapping lines (if multiple weeks) can get messy; doesn’t emphasize individual day values.

**Side-by-side comparison (right panel):**
- **Advantages:** Makes it easier to directly compare individual days (e.g., which days had more visitors week to week).
- **Disadvantages:** Less clear on trends across days; can feel cluttered with too many data series.

**Takeaway:** Use line charts when showing *patterns over time*, and use discrete comparison plots when *specific day-to-day differences* are the focus.

```python

import matplotlib.pyplot as plt

# Chapter 8-inspired SWD plot (minimal design, clear labeling)
channels = ['Online', 'In-store', 'Phone', 'Events']
sales = [55, 30, 10, 5]

plt.figure(figsize=(8,5))
bars = plt.bar(channels, sales, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

# Title & axis labels
plt.title("Sales by Channel", fontsize=16, fontweight='bold')
plt.ylabel("Sales (%)", fontsize=12)

# Remove clutter
for spine in ['top','right']:
    plt.gca().spines[spine].set_visible(False)

# Add direct labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f"{height}%", 
ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.ylim(0, 65)
plt.tight_layout()
plt.show()
```


### 📊 Storytelling With Data Plot

This plot is modeled after the redesign examples in Chapter 8 of *Storytelling With Data* (pp. 187–205).  
- **Why it works:** It uses direct labeling (values on bars), muted colors, and removes unnecessary lines and clutter.  
- **What it shows:** A simple breakdown of sales by channel, with a clear takeaway that **Online** sales dominate.
