```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
df = pd.DataFrame({'ser1': time_series, 'ser2': time_series_2}, index = datearray)
df.plot.area()
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 86-98 of the Storytelling With Data book as best you can.  (The second half of chapter three).  You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

# Data for the chart
categories = ['A', 'B', 'C', 'D']
values = [75, 82, 43, 67]
colors = ['#cccccc', '#cccccc', '#ff7f0e', '#cccccc']  # Emphasize Category C

# Create the figure and axis
fig, ax = plt.subplots(figsize=(6, 4))

# Plot the bars
bars = ax.bar(categories, values, color=colors)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 5), textcoords="offset points", ha='center', va='bottom')

# Add chart title and subtitle
plt.title("Category C Had the Lowest Value", fontsize=14)
ax.set_ylabel("Value")

# Remove unnecessary spines and ticks
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```

This bar chart follows principles from *Storytelling With Data*, specifically focusing on emphasis through color and simplification. Category C is highlighted to direct the viewerâ€™s attention, and gridlines are used minimally to supportâ€”not distract fromâ€”the insight.
