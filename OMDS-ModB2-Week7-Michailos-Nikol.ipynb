To help the viewer focus on the last three months of data (OctDec 2023), I used preattentive attributes like color and shape. Specifically, I made the earlier months gray and highlighted the final three months in red with markers. This draws attention immediately to the most recent trend.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate data
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
daterange = pd.date_range(start='2023-01-01', end='2024-01-01', freq='M')

# Create DataFrame
df = pd.DataFrame({'Date': daterange, 'Nitrate': time_series})

# Define cutoff for last 3 months
last_3_months = df['Date'] >= '2023-10-01'
df['Highlight'] = last_3_months

# Plot
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Date', y='Nitrate', color='gray', linewidth=2, label='All Data')
sns.lineplot(data=df[df['Highlight']], x='Date', y='Nitrate', color='red', linewidth=2.5, marker='o', label='Last 3 Months')

plt.title("Nitrate Level (% of eutrophic value)")
plt.xlabel("Year and Month")
plt.ylabel("Nitrate level")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dummy data
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    '2023': [65, 75, 45, 30, 90],
    '2024': [70, 60, 55, 35, 95]
}
df = pd.DataFrame(data)

# Plot setup
fig, ax = plt.subplots(figsize=(8, 6))

# Plot lines
for i in range(len(df)):
    color = 'red' if df['Category'][i] == 'E' else 'gray'
    linewidth = 2.5 if df['Category'][i] == 'E' else 1
    ax.plot(['2023', '2024'], [df['2023'][i], df['2024'][i]], 
            marker='o', color=color, linewidth=linewidth)
    ax.text('2023', df['2023'][i], df['Category'][i], ha='right', va='center', fontsize=10)
    ax.text('2024', df['2024'][i], df['Category'][i], ha='left', va='center', fontsize=10)

# Style
ax.set_title("Change in Category Scores from 2023 to 2024", fontsize=14)
ax.set_xlim(['2023', '2024'])
ax.set_xticks(['2023', '2024'])
ax.set_ylabel("Score")
plt.tight_layout()
plt.show()
```
